```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as scint 
nstep=20

#Energy range
ene=np.linspace(0.01,0.1,nstep)
t=np.zeros(nstep,float)
#Function to calculate gamma for given energy
def fint(en):
    xout= 1.0/en -1.0      # this would be a
    x = np.linspace(0,xout,1000)
    #Potential
    v= 1.0/(1.0+x)
    q=np.sqrt(abs(2*v-2*en))
    gam= scint.trapezoid(q,x)  # gamma integral
    return gam

#Loop to calculate transmission coefficient over energy range
for i in range(nstep):
    t[i]=np.exp(-2.0*fint(ene[i]))
#Plotting
plt.ylabel("Transmission Coefficient")
plt.xlabel("Energy")
plt.yscale("log")
plt.xscale("log")
plt.plot(ene,t,marker='o')
plt.show()

```

Energy of incident wave changes by one order of magnitude. 

Transmission coefficient changes by almost 14 orders of magnitude
