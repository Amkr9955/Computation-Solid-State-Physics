```python
# Band structure along \Gamma-X-M-\Gamma-R-X
import numpy as np
import matplotlib.pyplot as plt

k = []
ene = []

def fn(kx, ky, kz):
    value = kx**2 + ky**2 + kz**2
    return value

idx = 0.0

#  -> X
for i in range(100):
    dk = np.pi / 100
    kx = i/100 * np.pi
    ky = 0.0
    kz = 0.0
    k.append(idx)
    idx = idx + dk
    ene.append(fn(kx, ky, kz))

# X -> M
for i in range(100):
    dk = np.pi / 100
    kx = np.pi
    ky = i/100 * np.pi
    kz = 0.0
    k.append(idx)
    idx = idx + dk
    ene.append(fn(kx, ky, kz))

# M -> 
for i in range(100):
    dk = np.sqrt(2) * np.pi / 100
    kx = np.pi - i/100 * np.pi
    ky = np.pi - i/100 * np.pi
    kz = 0.0
    k.append(idx)
    idx = idx + dk
    ene.append(fn(kx, ky, kz))

#  -> R
for i in range(100):
    dk = np.sqrt(3) * np.pi / 100
    kx = i/100 * np.pi
    ky = i/100 * np.pi
    kz = i/100 * np.pi
    k.append(idx)
    idx = idx + dk
    ene.append(fn(kx, ky, kz))

# R -> X
for i in range(100):
    dk = np.sqrt(2) * np.pi / 100
    kx = np.pi
    ky = np.pi - i/100 * np.pi
    kz = np.pi - i/100 * np.pi
    k.append(idx)
    idx = idx + dk
    ene.append(fn(kx, ky, kz))

plt.ylabel("Energy")
plt.plot(k, ene)
plt.show()
```
